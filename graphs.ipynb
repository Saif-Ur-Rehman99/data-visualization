```python
from matplotlib import pyplot as plt
import numpy as np
```

Line Graph

```python
x = [1,2,3,4,5] 
y1 = [2,3,5,7,4] 
y2 = [1,4,6,2,7] 
plt.xlabel("Supply", fontsize=15) 
plt.ylabel("Demand", fontsize=15) 
plt.title("Line Graph", fontsize=20) 
plt.plot(x,y1, color='grey') 
plt.plot(x,y2, color='pink') 
plt.grid() 
plt.show() 
```

Bar Graph (Vertical)

```python
x = ['Python', 'Java', 'Swift', 'C++', 'SQL', 'C'] 
y = [85, 80, 52, 70, 80, 50] 

plt.bar(x,y, color='pink',edgecolor='black', align='edge', label='lang') 
plt.legend() # Only, when we use label parameter 
plt.xlabel("Languages") 
plt.ylabel("Users") 
plt.title("Popularity") 
plt.show() 
```

Multiple Bar Graph (Vertical) 

```python

x = ['Python', 'Java', 'Swift', 'C++', 'SQL', 'C'] 
y1 = [94, 75, 80, 64, 70, 45] #ML 
y2 = [90, 85, 70, 70, 90, 50] #WEBAPP 
  

width = 0.4 
p = np.arange(len(x)) 
p1 = [j+width for j in p] 

plt.bar(p,y1, color='yellow',edgecolor='black', width=0.4, label='ML') 
plt.bar(p1,y2, color='red',edgecolor='black', width=0.4, label='WebApp') 
plt.legend() # Only, when we use label parameter 

plt.xlabel("Languages") 
plt.ylabel("Users") 
plt.title("Popularity") 
plt.xticks(p+width/2, x, rotation=20) # for python java swift etc 
plt.show() 

# edgecolor = border-color-name  
# align = "edge" or "center" 
# width = width of bar (0.1 - 1.0) 
# linewidth = width of border (10) 
# linestyle = ":"(dotted), "--"(dashed), "_."(dotted and dashed) 
# alpha = 0.4 (opacity faded) 
# label='label's name (plt.legend() must used) 
```
